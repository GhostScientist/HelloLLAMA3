This is my first Colab notebook. I'm going to try and use what I learned recently at ODSC from Jon Krohn to try some NLP tasks using the latest LLAMA 3 model from Meta.

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

from huggingface_hub import login
from google.colab import userdata

login(userdata.get('HF_TOKEN'))

pipe = pipeline("text-generation", model="meta-llama/Meta-Llama-3-8B")
```

```python
prompt = "Can you repeat back to me only the following words: Hello, my name is Carl Wheezer, and I love Llamas!"
output = pipe(prompt, max_new_tokens=20, num_return_sequences=1)
```

```python
print(output[0]['generated_text'])
```
